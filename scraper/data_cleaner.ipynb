```python
import pandas as pd

data = pd.read_csv("./facebook_1.csv")
```

```python
data = data.drop(["x193iq5w 6"], axis=1)
data
```

```python
concatenated = []
for _, row in data.iterrows():
    concatenated.append(" ".join([val for val in row if pd.notna(val) and val != ""]))
concatenated = [s for s in concatenated if s != ""]
```

```python
with open("output.txt", "w") as f:
    for element in concatenated:
        f.write(str(element) + "\n")
```
